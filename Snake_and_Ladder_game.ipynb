```python
import random

# Function to roll the dice
def roll_dice():
    return random.randint(1, 6)

# Function to check if a ladder is encountered
def check_ladder(position):
    ladder ={1:38, 4:14, 9:31, 21:42, 28:84, 51:67, 71:91, 80:100}
    if position in ladder:
         print("yay! you climed a ladder  ",ladder[position])
         return ladder[position]
         
    else:
        return position

# Function to check if a snake is encountered
def check_snake(position):
    snake = {17:7,54:34,62:19,64:60,87:24,93:73,95:75,98:79}
    if position in snake:
         print("Oops! snake bite!",snake[position])
         return snake[position]
    else:
        return position

# Function to play the game
def play_game():
    # Initial positions for both players
    player1_position = 0
    player2_position = 0
    
    # Variable to keep track of player
    player = 1
    
    # Loop until either player reaches or crosses 100
    while player1_position < 100 and player2_position < 100:
        if player == 1:
            input("Player 1, press Enter to roll the dice:")
            dice_roll = roll_dice()
            print("Player 1 rolled a", dice_roll)
            if dice_roll == 6:
                print("Player 1 gets an extra chance!")
            player1_position += dice_roll
            player1_position = check_ladder(player1_position)
            player1_position = check_snake(player1_position)
            print("Player 1 is now at position", player1_position)
            if dice_roll != 6:
                 player = 2   #here we are intializing the player =2 for player 2 to get the chance to play
        else:
            input("Player 2, press Enter to roll the dice:")
            dice_roll = roll_dice()
            print("Player 2 rolled a", dice_roll)
            if dice_roll == 6:
                print("Player 2 gets an extra chance!")
            player2_position += dice_roll
            player2_position = check_ladder(player2_position)
            player2_position = check_snake(player2_position)
            print("Player 2 is now at position", player2_position)
            if dice_roll != 6:
                  player = 1 
    
    # Check the winner
    if player1_position >= 100:
        print("Player 1 wins!")
    else:
        print("Player 2 wins!")

# Start the game when a player gets 6
while True:
    input("Press Enter to roll the dice and start the game:")
    dice_roll = roll_dice()
    print("You rolled a", dice_roll)
    if dice_roll == 6:
        print("Game starting!")
        play_game()
        break
    else:
        print("You need to roll a 6 to start the game.")
```
