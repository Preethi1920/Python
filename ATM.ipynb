```python
class ATM():
    def __init__(self):
        self.__pin=0
        self.balance=0
        self.info()
        
        
    def info(self):
        input_value=int(input("""Select one of these :
                               1: Create the pin for new user.
                               2: Check balance
                               3: Withdrawl
                               4: Deposit
                               5: Mini statement
                               6. Change Pin
                               7: Exit\n"""))
        def create_pin():
            if self.__pin==0:
                new_pin1=int(input("Enter your new pin: "))
                new_pin2=int(input("Enter your new pin again: "))
                if new_pin1==new_pin2:
                    self.__pin=new_pin2
                    print("Pin updated successfully.")
                else:
                    print("Pin mismatch! ")
            else:
                print("Pin already exists!")
            
            x=input("Press 'b' to go back to main menu or 'e' to exit :")
            if x=='b':
                self.info()
            else:
                pass
                      
                
        def deposit():
            check=int(input("Enter your pin: "))
            if check==self.__pin:
                bal=int(input("Enter the amount you want to deposit: "))
                notes_100=int(input("Enter the no of 100 notes: "))
                notes_200=int(input("Enter the no of 200 notes: "))
                notes_500=int(input("Enter the no of 500 notes: "))
                notes_2000=int(input("Enter the no of 2000 notes: "))
                if (100*notes_100+200*notes_200+500*notes_500+2000*notes_2000)==bal:
                    self.balance=bal
                    print("Deposit successful !")
                else:
                    print("Notes mismatch! ")
                
            else:
                print("Incorrect pin entered! ")
        
        def withdrawl():
            check=int(input("Enter your pin: "))
            if check==self.__pin:
                how_much_wd=int(input("Enter the amount you want to withdraw: "))
                if how_much_wd<=self.balance:
                    self.balance-=how_much_wd
                else:
                    print("Insufficient balance! ")
            else:
                print("Incorrect pin entered! ")
                    
        def balance():
            check=int(input("Enter your pin: "))
            if check==self.__pin:
                print(f"Your account balance is : {self.balance}")
            else:
                print("Incorrect pin entered! ")
            x=input("Press 'b' to go back to main menu or 'e' to exit :")
            if x=='b':
                self.info()
            else:
                pass
                
        def change_pin():
            check=int(input("Enter your previous pin: "))
            if check==self.__pin:
                new_pin1=int(input("Enter your 4 digit new pin : "))
                new_pin2=int(input("Enter your 4 digit new pin again : "))
                if new_pin1==new_pin2:
                    self.__pin=new_pin1
                    print("New pin updated successfully! ")
                else: 
                    print("Pin mismatch !")
            else:
                print("Incorrect pin entered! ")
            x=input("Press 'b' to go back to main menu or 'e' to exit :")
            if x=='b':
                self.info()
            else:
                pass
            
        if input_value==1:
            create_pin()
        elif input_value==6:
            change_pin()
        elif input_value==4:
            deposit()
        elif input_value==3:
            withdrawl()
        elif input_value==2:
            balance()
        else:
            pass
```
